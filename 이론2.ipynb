# 1. 엔터티(entity)-현실세계에서 독립적으로 식별 가능한 객체나 사물
-엔터티는 대상(instance-행)로 이루어진 집합
## 예시
  
속성=컬럼, 인스턴스=행  
## 엔터티의 특징
-행들은 식별가능해야하므로 유일한 식별자가 필요(동명이인이 있을수도 있으므로)  
-필요한 정보여야함  
-인스턴스들의 집합(최소 2개 이상)  
-반드시 속성을 가짐(최소 2개 이상의 컬럼)  
-사용되어야함-필요성이 없으면 고립 엔터티는 제거 필요  
-다른 엔터티와 최소 1개 이상의 관계 성립이 필요-ex) 부모와 자식

# 1-1 엔터티의 분류
## 1. 유형과 무형
-유형엔터티: 물리적 형태 존재(사원,물품)  
-개념엔터티: 물리적 형태 없음(조직, 보험상품)  
-사건엔터티: 업무 수행에 따라 발생(주문, 청구, 미납)
## 2. 발생 시점
-기본 엔터티: 원래 존재하던 정보(사원, 부서, 고객, 상품)->자신의 고유한 주식별자를 가짐  
-중심엔터티: 기본엔터티로 부터 발생(계약,사고,청구,주문,매출->제일 중심적인 역할을 하며 다량의 데이터 발생  
-행위엔터티: 2개 이상의 부모엔터티로부터 발생(주문[고객과 상품으로 부터 파생],사원변경이력)->자주 내용이 바뀜

# 2. 속성(attribute): 업무에서 필요로 하는 고유한 성질, 특징을 의미->컬럼으로 표현할 수 있는 단위
## 특징
-더 이상 분리되지 않음  
-인스턴스의 구성요소임  
-한개의 엔터티에는 최속 2개싱상의 인스턴스, 2개이상의 속성 필요  
-한 개의 속성은 1개의 속성값만 가짐(컬럼하나에 값 하나)->속성의 원자성
-정해진 주식별자(학번 등)에 함수적 종속성을 가짐
## 함수적 종속성-한 속성값이 다른 속성값에 종속적인 관계를 가짐
-1. 완전함수적 종속: 특정 컬럼이 기본키에 완전 종속-ex)제품가격과 나의 돈으로 살수있는 개수 컬럼이 결정   
-2. 부분함수적 종속: 일부에 대한 종속-ex) 어떤 학생이 수학 과목을 들어도 학생에 상관없이 수학과목 강사는 해당 과목 컬럼에 의해 정해져있음
## 속성의 분류-1: 속성의 특징에 따른
-1. 기본속성: 업무로 부터 추출(원금, 예치기간)  
-2. 설계속성: 새로 만들어지거나 기존으로 부터 변형하여 만들어진 속성(상품코드, 지점코드등 기존에 있었지만 구분하기 위해)  
-3. 파생속성: 다른 속성에 의해 만들어짐(합계, 평균, 이자)
## 속성의 분류-2: 엔터티 구성방식에 따른
-1. PK(primary key): 인스턴스를 식별할 수 있는 속성  
-2. FK(foreign key): 다른 엔터티와의 관계에서 포함된 속성(참조)  
-3. 일반속성: 엔터티에 포함되지만 PK/FK에는 포함되지 않는 속성
## 속성의 분류-3: 분해여부에 따른
-1. 단일속성: 하나의 의미로 구성된 경우(회원 ID, 이름)  
-2. 복합속성: 여러개의 의미로 구성(주소-시+구+동)  
-3. 다중값속성: 속성에 여러개 값을 가질 수 있는 경우(상품 리스트나 한 사람이 가입 한 여러 상품)

# 3. 관계(relationship)-엔터티간의 관계
## 관계의 종류
-1. 존재적 관계: 한 엔터티의 존재가 다른 엔터티의 존재에 영향(부서가 사라지면 사원 엔터티의 존재에도 영향)  
-2. 행위적 관계: 고객 엔터티의 행동에 의해 주문 엔터티가 발생
## 관계의 구성
-1. 관계명: 이름짓기  
-2. 차수(cardinality): 한 엔터티의 인스턴스가 다른 엔터티의 인스턴스와 어떻게 연결되는지를 표현  
-- 1대1 관계: 완전 1대1(사원은 반드시 소속 부서가 있어야함) vs 선택적 1대1관계(아직 발령전이라 없을 수 있음)  
-- 1대N관계: 하나의 부서에 여러명의 사원  
-- M대 N관계: 한 학생이 여러 강의 수강 가능+ 그 강의도 여러 학생 보유 가능  
** 이때 인스턴스끼리의 관계는 페어링이라함(학생이 강의를 듣고 A의 학점을 받았다)  
-3. 선택성: 필요한지 선택적인지
## 강한 개체 vs 약한 개체
-강한개체: 독립적으로 존재 가능한 엔터티(고객과 계좌중 '고객')  
-약한개체: 독립적으로 존대할 수 없는 엔터티(고객과 계좌중 '계좌')
## JOIN

# 4. 식별자-여러개의 속성중 엔터티를 대표할 수 있는 속성(예시-id)
-하나의 유일한 식별자가 존재해야함(primary key)
## 주식별자 특징
-유일성  
-최소성(학번으로 충분, 학번+이름은 투머치)  
-불변성  
-존재성(NULL은 X)  
## 식별자 분류
  
ex) 주민번호또한 고유한 번호지만 보안상 쓰지 않음->보조식별자


  
본질식별자: 부서번호 vs 인조식별자: 인위적으로 고유한 값 만들어 사용  
인조식별자 단점-중복데이터 발생+불필요한 인덱스 생성->성능저하
## 어떤 엔터티에서는 식별자일 수 있지만 다른 엔터티에서는 식별자가 아닐 수 있음

## 식별자 표기법
-1. IE표기법에서는 네모로  
  
-2. Barker표기법에서는 #으로  

## 주식별자 기준
-1. 자주 이용되는 속성  
-2. 명칭이나 내역등과 같은 '이름'은 피함(부서명 보다는 부서 코드)  
-3. 속성 수를 최대한 적게-> 주식별자가 여러개면 차라리 새로운 인조식별자를 생성하는게 이후 조인등의 성능저하 감소
## 식별관계와 비식별관계
-1. 식별관계: 하나의 엔터티의 기본키를 다른 엔터티가 기본키의 하나로 공유(ERD에서는 실선으로 표시)  
-2. 비식별관계: 강한개체의 기본키를 다른 엔터티에서는 기본키가 아닌 일반 속성으로 관계를 가짐(ERD에서는 점선으로 표시)  
ex) 부서는 부서코드가 기본키고 그에 해당하는 약한 개체인 사원객체에는 부서코드가 있지만 사원코드를 주식별자로 사용
## key의 종류(논리 모델에서는 식별자라 부르지만 물리 모델링으로 넘어가면 KEY라고 부름)
-1. 기본키(primary key): 엔터티 대표 가능  
-2. 후보키(candidate key): 유일성과 최소성을 가지며 기본키가 될 수 있음-> 기본키가 결정되면 나머지 후보키는 대체키라고 부르게됨  
-3. 대체기: 여러 후보키중 기본키가 아닌 키  
-4. 슈퍼키: 유일성은 만족하지만 최소성은 만족X(학번+이름)  
-5. 외래키: 다른 테이블의 기본키를 참조하는 키(존재하는 값만 입력가능-R의 factor처럼)
