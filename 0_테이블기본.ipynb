# 관계형 데이터베이스: 데이터를 행과 열이 있는 테이블 형식으로 저장 -> 테이블을 연결하여 다양한 데이터 포인트 간의 상호 연결에 대한 심층적인 인사이트를 얻을 수 있음
# 관계형 데이터베이스 구성요소3
### 계정+테이블+스키마(테이블이 어떤 구성과 정보를 가지고 있는지에 대한 기본 구조)
# 이론
-하나의 테이블은 반드시 하나의 계정 소유여야함  
-테이블 명은 중복될 수 없지만 소유자가 다른 경우 같은 이름으로 생성가능
## SQL(structured query language)
- 관계형 데이터베이스에서 쓰이는 모든 언어
- 데이터정의(DDL), 데이터조작(DML), 데이터 제어 언어(DCL)등으로 구분
## query:데이터베이스에게 특정 정보를 보여달라는 사용자의 요청

# 기본 특징
### 기본구조
### 여러가지 db들 중 'hi' db에 들어가면 테이블이 있음->+하면 여러 테이블들이 나옴->이중하나골라+하면 각 컬럼들이 나옴
-sql문 실행할때는 ctrl+enter  
-sql문은 대소문자를 구분하지 않음  
-sql문은 한 줄 또는 여러줄에 입력가능but 가독성을 위해 한줄에 하나씩  
### 키워드(select, from where, group by, having, order by)는 약어존재X + 여러줄로입력X(sel띄고 다음줄에ect불가)
-가독성을 높이기 위해 들여쓰기사용  
### 여러sql문을 사용할때는 끝에';'사용하여 구분
-열 머리글=sql문 실행했을때 나오는 결과에서 맨위 컬럼부분-기본값은 왼쪽정렬+대문자

-산술식: +(더하기), -(빼기), *(곱하기), /(나누기)  
  
연산자 우선 순위: 곱셈,나눗셈이 덧셈,뺄셈보다 우선 but ()사용시 ()안에있는것을 우선사용
