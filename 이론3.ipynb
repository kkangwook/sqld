# 5. 정규화(DB Normalization)-엔터티(테이블)를 분해하는 과정
### 원하는 정보만을 가져와 일관성, 최소한의 중복, 데이터 유연성을 확보하는 과정
## 이상현상
-정규화를 하지않아 발생하는 현상  
-너무많은 데이터에 의해 데이터 관리하는 중 오류나는 현상
## 정규화 단계
-1. 제 1정규화(1NF): 컬럼이 원자성을 갖도록 테이블을 분해하는 단계   
  
-2. 제 2 정규화(2NF): 테이블에 완전 함수 종속을 만들도록 테이블을 분해(PK가 2개 이상일때 발생->분리)  
  
강의실은 강의명에 의해서만 결정되므로 부분함수종속  
->강의실을 빼서 따로->완전함수종속  
또한 성적은 학생번호와 강의명에 의해 결정되므로 완정함수종속에 만족  
-3. 제 3정규화(3NF): 이행적 종속을 없애는 과정   
이행적 종속: A->B, B->C의 관계가 성립할때 A->C가 성립되는 경우를 말함  
따라서 (A,B)와 (B,C)로 분리해야함  
  
이유: 만약 상품명을 변경하면 가격도 같이 수동으로 변경해 줘야 해서
## 일반적으로 3단계까지만 진행
-4.제4정규화: 여러 컬럼이 하나의 컬럼을 종속시키는 경우 분해하여 다중값 종속성을 제거  
-5. 제 5정규화: 조인에 의해서 종속성이 발생되는 경우 분해
## 반정규화(역정규화)
너무 과도하게 정규화하면 조회(select)속드는 향상되지만 join등의 수행은 느려짐  
-> 정규화의 반대방향으로 되돌아감(중복등을 일부 허용)  
특정 범위가 자주 사용되고 다량의 범위를 자주 처리하는 등의 경우 반정규화 하기도 함

# 7. NULL-아직 정해지지 않은 값을 의미
0이나 ''와는 다름
## NULL 특성
-1. NULL을 포함한 연산결과는 항상 NULL  
  
NULL을 무시하고 싶을때->NVL함수 사용  
  
NVL(COMM,0)은 NULL일시 0의 값으로 대체하겠다는 것을 의미  
-2. 그룹함수는 NULL을 제외하고 연산을 수행한다(SUM,AVG,MIN,MAX등)  
이때 count(컬럼)은 NULL이 아닌값만 세지만 count(*)는 NULL을 포함한 모든 행의 수를 리턴  
이때도 마찬가지로 NULL포함하고 싶은면 sum(NVL(comm,0))으로 NULL값들도 포함
