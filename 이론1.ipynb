# 1.모델링-현실세계의 데이터를 추상적이고 구조화된 형태로 표현하는 과정
## 특징
-단순화: 핵심요소에 집중, 불필요한 세부사항 제거  
-추상화: 현실세계를 일정한 형식에 맞추어 간략하게 표현하는 과정  
-명확화: 애매모호함을 제거하고 정확하게 현상을 기술  
## 모델링 3가지 관점
-데이터관점: 데이터 그 자체  
-프로세스관점: 업무에 더 치중, 시스템이 어떤 잡업을 수행하는지  
-데이터와 프로세스관점: 데이터+프로세스 결합하여 특정 프로세스가 어떤 데이터를 사용하는지 시스템의 전반적인 동작을 이해
## 모델링 유의점
-중복: 같은 정보 없도록 설계  
-비유연성: 사소한 업무 변화에 대해 작은 모델 변경이 되지 않도록  
-비일관성: 정보가 서로 모순되거나 상반되면 안됨(물론 비일관성은 발생할 수 있음)
## 데이터 모델링 3가지 요소
-대상(entity): 업무가 관리하고자하는 대상  
-속성(attribute): 대상들이 갖는 속성  
-관계(relationship): 대상들 간의 관계
## 모델링의 3단계
-1.개념적 모델링: 추상화 정도가 가장 큼, 핵심 엔터티를 추출, 엔터티들과의 관계들을 표현하기 위한 ERD작성  
-2.논리적 모델링: 개념적 모델링을 기반으로 세부속성(이름, 주민번호 등), 식별자, 관계등을 표현->데이터 정규화 수행(어떻게 쪼갤지)  
-3.물리적 모델링: 논리모델링이 끝나면 이를 토대로 직접 물리적으로 생성(테이블,컬럼생성+성능,저장구조,보안성 등 고려)-추상화 수준이 가장 낮음

# 2. 스키마-모든 객체(메타데이터)들의 정보의 집합
# 객체=테이블, 인덱스, 제약조건등
## 스키마의 3단계 구조
-외부스키마: 사용자가 보는 관점  
-개념스키마: 논리적 모델링에서 만들어진 스키마->논리적 구조를 정의(데이터 베이스의 개체,속성,관계,데이터타입 등을 정의)  
-내부스키마: 물리적으로 어떻게 저장되는지를 정의->데이터의 저장구조,컬럼,인덱스등을 정의
## 스키마의 독립성-한쪽이 변경되더라도 다른쪽이 변경되면 안된다
-논리적 독립성: 개념스키마가 변경되어도 프로그램에 영향을 주지 않음  
-물리적 독립성: 물리적 구조가 변경되어도 외부/개념 스키마에 영향을 주지 않음
